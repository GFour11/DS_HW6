```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
df_2d = pd.read_csv('data_2d.csv')
df_2d.head()
```

```python
values = df_2d.values

n_clusters = list(range(1, 10))
distortions = []
for i in n_clusters:
  km = KMeans(n_clusters=i, random_state=0, n_init="auto").fit(values)
  distortions.append(km.inertia_)

plt.plot(n_clusters, distortions)
plt.title('Elbow method')
plt.xlabel('Clusters')
plt.ylabel('Distortions') 
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(values)
labels = kmeans.labels_

plt.scatter(values[:,1], values[:,2], c=labels)
plt.title('Clusters')
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
mnist = pd.read_csv('mnist.csv')
```

```python
standart_scaler = StandardScaler()
scale = standart_scaler.fit_transform(mnist)


pca = PCA(n_components=2)
new_scale = pca.fit_transform(scale)


distortions = []

for i in range (1,11):
    km = KMeans(n_clusters=i, random_state=0, n_init='auto').fit(new_scale)
    distortions.append(km.inertia_)

plt.plot(range(1,11),distortions)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(new_scale)
labels = kmeans.labels_

plt.scatter(new_scale[:,0], new_scale[:,1], c=labels)
plt.title('Clusters')
```
